```python
#######################
###FOUNDATIONS SETUP###
#######################

import random

#random.seed(10)
number = str(random.randrange(1000, 9999))

greetings = "Hello there!"
text = """I've generated a random 4 digit number for you.
The number is between 1000 and 9999.
If you get all numbers correct and in the right places, you win!
If not, you will get one bull for each number that is on the right spot
and/or one cow for each number that is in the generated number but you didnt
put it on the right spot.
Be careful, count matters! For example: generated nr = 1111, your nr = 4321 --> 0 bulls, 1 cow
Let's play a bulls and cows game.
I will keep the number of guesses it took to find the number."""
separator = "-" * (len(greetings) * 3 )
game_running = True
bulls = 0
cows = 0
used_guesses = list()

print(f"""
{separator}
{greetings}
{separator}
{text}
{separator}"""
, sep="\n")

while game_running:
    user_input = str(input("Enter a 4-digit number: "))

######################
###GUESS EVALUATION###
######################

    if user_input == "q":
        print("You decided to end the game. Goodbye ...")
        game_running = False
        break
    elif user_input != "q" and (len(user_input) != 4 or not user_input.isnumeric() or int(user_input) < 1000):
        print(f">>> {user_input}")
        print("Entered number doesnt have 4 digits, doesnt contain only numbers or is not between 1000 and 9999!")
        print(separator)
        continue
    elif user_input in used_guesses:
        print(f"You have already tried {user_input} - try again something different!")
        print(separator)
        continue
    elif str(''.join(user_input)) == str(''.join(number)):
        used_guesses.append(user_input)
        if len(used_guesses) == 1:

            print(f"CONGRATS! You guessed the number: {user_input}! It took you only {len(used_guesses)} guess!")
        else:
            print(f"CONGRATS! You guessed the number: {user_input}! It took you only {len(used_guesses)} guesses!")
        print(separator)
        if len(used_guesses) < 9:
            print("Damn, ur goooooood!")
        elif len(used_guesses) < 17:
            print("Well done, thats not bad at all...")
        elif len(used_guesses) < 26:
            print("There are other things you are good at, Im sure")
        elif len(used_guesses) >= 26:
            print("You better practice")
        break
    used_guesses.append(user_input)

################################
###COWS AND BULLS CALCULATION###
################################

    user_input = list(user_input)
    number = list(number)
    bull_list = list()
    cow_list = list()
    rest_user_input = list()
    rest_number = list()

    for i in range(len(user_input)):
        if user_input[i] == number[i]:
            bull_list.append(int(user_input[i]))
        else:
            rest_user_input.append(user_input[i])
            rest_number.append(number[i])
    for i in range(len(rest_user_input)):
        if rest_user_input[i] in rest_number:
            rest_number.remove(str(rest_user_input[i]))
            cow_list.append(str(rest_user_input[i]))

###############################
###CALCULATIONN OUTPUT PRINT###
###############################
            
    print(f">>> {''.join(user_input)}")
    if len(bull_list) == 1:
        print(f"{len(bull_list)} bull,", end=' ')
    else:
        print(f"{len(bull_list)} bulls,", end=' ')
    if len(cow_list) == 1:
        print(f"{len(cow_list)} cow")
    else:
        print(f"{len(cow_list)} cows")
    print(separator)

###############################
###if playing:
###    have fun
###elif assessing:
###    feel appreciated :)
###else:
###    have a great day!
###############################
```
