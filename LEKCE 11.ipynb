<h3> Formát JSON </h3>

- .json
- např. jako:
{
    "jmeno": "Chuck Norris",
    "neuspech": null,
    "kliky": "vsechny",
    "konkurence": false,
}
- podobný dictu, trošku jiná pravidla (str --> strign, True --> true, False --> *false, None --> null)
- účel především v přenosu dat, záměrně zjednodušený, nestandardní datový typ pro Python (zejména v eshopech)
- aka JavaScript Objet Notation
- není standardní --> potřeba built-in knihovny json
- důležité metody
    - json.dump(m, n) ... zapíše objekt do JSONu
    - json.dumps(m) ... zapíše objekt do stringu
    - json.load(m) ... načte data z JSONu
    - json.loads(m) ... načte data ze stringu
    - kde m je jméno objetku, n jméno souboru
- zapisování objektu JSON = serializace
    - např. vezmu slovník, kde jsou None a False hodnoty někde
    - použiju fci s metodou ... json.dumps(slovnik)
    - převede None na null a False na *false a nonASCII znaky "zkomolí" na \kombinaci_znaku (např. Ł na \u0141)
    - dumps X dump --> dumps zapíše do stringu, dump (bez s) podobně jako texťák - otevřu v nějakém režimu, zapíšu, zavřu whatever
    - u dumps je jeden argument (co dumpuju), u dump 2 (co dumpuju, kam dumpuju)
    - dump: podobně jako u texťáku mohou použití kontextového manažera, write, read a tak
- čtení objektu JSON = deserializace
    - když použiju dumps, tak mi to vrátí string
    - takže když tam loaduju slovník, tak ho ale chci zpátky jako slovník občas --> fce/metoda .loads --> např. json_vystup= json.loads(json_vystup)
    - pro čtení souboru opět variace --> metoda .load(jmeno_souboru)
- když chci zapisovat znaky mimo ASCII tabulku, použuji argument ensure_ascii
    - např. vystup_s_jsonem = json.dumps(chuckuv_slovnik, ensure_ascii=False)
- když odsazovat, dám volitelné argument indendt=
    - např. vystup_s_jsonem = json.dumps(chuckuv_slovnik, ensure_ascii=False, indent=4)
- seřazení klíčů --> sort_key = True
    - např. vystup_s_jsonem = json.dumps(chuckuv_slovnik, ensure_ascii=False, indent=4, sort_keys=True)
    - defaultně je False

```python
import json

data = [
    {
        "id":1,
        "jmeno":"Marek",
        "barva": [
           "červená",
           "zelená"
        ]
     },
     {
        "id":2,
        "jmeno":"Bob",
        "barva": [
           "růžová",
           "žlutá"
        ]
     }
  ]

vysledek = list()

def najdi_zadane_klice(jmeno_souboru):
   global vysledek
   obsah_jsonu = 0
   with open(f"{jmeno_souboru}", mode="r") as json_soubor:
      obsah_jsonu = json.load(json_soubor)
      for i in obsah_jsonu:
         if "jmeno" in i.keys():
            vysledek.append(i.get("jmeno"))
   return vysledek

print(najdi_zadane_klice("udaje.json"))
```

<h3> Cvičení 2: Seřaď klíče </h3>

```python
import json

jmeno_souboru = "serazene.json"
data = {'4': 5, '6': 7, '1': 3, '2': 4}
vysledek = 0

def zapis_serazene_klice(jmeno_souboru, data):
    with open(f"{jmeno_souboru}", mode="w") as json_file:
        json.dump(data, json_file, sort_keys=True)

def vypis_obsah_souboru(jmeno_souboru):
    with open(f"{jmeno_souboru}", mode="r") as json_file:
        vysledek = json.load(json_file)
    return vysledek

zapis_serazene_klice(jmeno_souboru, data)
vypis_obsah_souboru(jmeno_souboru)
```

<h3> Formát CSV </h3>

- tabulková data, oproti xlsx v textové podobě
- CSV = comma separated values
- dialekt aka formátovací pravidla CSV
    - oddělení buněk pomocí středníku ;
    - oddělení řádků
    - apod. jako jsou escape znaky, použití uvozovek aj.
- opět potřeba built-in knihovny --> import csv
- hlavně tyto metody (základ): csv
    - .writer(m) ... převede zadaná dato do objektu m
    - .reader(m) ... vytvoří iterovatelný objekt s jednotlivými řádky z objektu m
    - .DictWriter(m) ... převede zadaná data do objektu m, pracuje s dict
    - .DictReader(m) ... vytvoří iterovatelný objekt s jednotlivými řádky z objektu m, pracuje s dict
    - objekt m = jméno objektu
- zápis nového CSV
    - podobně jako pro txt s open
    - pak je potřeby vytvořit special objekt writer s jehož pomocí lze zapisovat data po řádcích
    - writer - první způsob zápise pomocí fce writer ... pro uložení datových typů list nebo tuple
    - writer je speciální datový typ ... _csv.writer
- defaultní dialekt
    - řádky nerozděleny konkrétním symbolem (newline="")
        - TOHLE CHCI DÁVAT, když mám třeba list listů --> jinak to háže prázdný řádek mezi něma
    - buňky odděleny čárkou (delimiter=",")
    - ... dle toho pak setupuju otevření v Excelu
- zápis jednoho řádku: zapisovac.writerow(data)
- víc řádků: zapisovac.writerows((data, data1, data2))
    - POZOR, očekává pouze 1 argument (!) --> proto řádky v listu/tuplu
- ... samozřejmě lze taky pracovat pomocí kontext manažeru: with open(nazev_souboru, mode) as nove_csv: zapisovac.writer(nove_csv) zapisovac.writerow(...)
- DictWriter: když mám předepsanou hlavičku a chci zapisovat pomocí dictu, kde klíč je sloupec, hodnota je hodnota
    - velká výhoda: VYTVOŘENÍ JMEN SLOUPCŮ dle klíčů
    - 1) otevřu soubor
    - 2) z klíčů nachystám hlavičku (zahlavi = slovnik.keys())
    - 3) nachystám zapisovač (zapisovac = csv.DictWriter(nazev_souboru, fieldnames=zahlavi))
    - 4) zapíšu záhlaví (zapisovac.writeheader())
    - 5) zapíšu data (zapisovac.writerow(data))
    - 6) zavřu
    - opět lze použít kontext manažer
- čtení existujícího CSV
    - zase open
    - místo zapisovače potřebuju "čteče" --> csv.reader
    - btw ho lze iterovat (for cyklus, výpis pomocí list/tuple)
    - čteč/čtenář je taky special typ csv.reader
    - 1) otevřu soubor (cteni_csv = open(nazev_souboru))
    - 2) uložím čtení (ctenar = csv.reader(cteni_csv))
    - 3) printnu čtení jako list/tuple (print(tuple(ctenar)))
    - 4) zavřu
    - opět kontext manažer possible
- DictReader
    - nevrací jako jeden list/tuple, plný listů
    - ale každý řádek = oddělený dict
    - AUTOMATICKY VEZME HEADER JAKO KLÍČE
    - opět kontext manažer viable cesta
- užitečné argumenty (volitelné)
    - delimiter ... definuju oddělovač, dávám do writeru jako argument (např. delimiter = ";")
    - dialect ... použití přednastavených dialektů (taky v zapisovači, např. dialect="excel-tab")

```python
import csv

hlavicka = ["jmeno", "prijmeni", "vek"]
osoba_1 = ["Matous", "Pokoj", "28"]
osoba_2 = ["Petr", "Svetr", "27"]

#otevřu
nove_csv = open("prvni_csv_soubor.csv", mode="w")
#vytvořím zapisovač/writer
zapisovac = csv.writer(nove_csv, delimiter=";")
#zapíšu hlavičku
zapisovac.writerow(hlavicka)
#a řádek 1 a 2
zapisovac.writerow(osoba_1)
zapisovac.writerow(osoba_2)
zapisovac.writerows(
    (
        osoba_1,
        osoba_2
    )
)
nove_csv.close()
```

```python
osoba_1 = {"jmeno": "Matous", "prijmeni": "Pokoj", "vek": "28"}
osoba_2 = {"jmeno": "Petr", "prijmeni": "Svetr", "vek": "27"}

nove_csv = open("druhy_csv_soubor.csv", mode="w")
# z klíčů udělám hlavičku
zahlavi = osoba_1.keys()
#nachystám zapisovač
zapisovac = csv.DictWriter(nove_csv, fieldnames=zahlavi, dialect="excel-tab")
#zapíšu záhlaví
zapisovac.writeheader()
#a oba údaje
zapisovac.writerow(osoba_1)
zapisovac.writerow(osoba_2)
#zavřu...
nove_csv.close()
```

```python
cteni_csv = open("prvni_csv_soubor.csv")
#vytvoření iterovatelného čtenáře
ctenar = csv.reader(cteni_csv)
#výpis pomocí tuple
print(tuple(ctenar))        #POZOR! - normálně by každý řádek byl ve stylu: ['jmeno', 'prijmeni' ...]
#zavřu                      ALE skrz volitelný delimiter (;) to vypadá takto
cteni_csv.close()
```

```python
cteni_csv = open("druhy_csv_soubor.csv")
#vytvořím čtenáře
ctenar = csv.DictReader(cteni_csv)
#vypíšu obsah csv pomocí tuple
print(tuple(ctenar))            #POZOR: tady by se to normálně zobrazilo jako list slovníků, kde se rovnou načtou klíče k hodnotám
#zavřu                          ALE skrze volitelný dialekt (viz výše) je to takovéto
cteni_csv.close()
```

<h3> Cvičení 3: Zapiš a přečti CSV </h3>

```python
import csv

data = [
    [10,'a1', 1],
    [12,'a2', 3],
    [14, 'a3', 5],
    [16, 'a4', 7],
    [18, 'a5', 9]
]

with open("nove.csv", mode="w", newline="") as nove_csv:
    zapisovac = csv.writer(nove_csv)
    zapisovac.writerows(data)

with open("nove.csv", mode="r") as nove_csv:
    ctenar = csv.reader(nove_csv)
    csv_data = list(ctenar)
    for i in csv_data:
        print(*i, sep=",")
```
