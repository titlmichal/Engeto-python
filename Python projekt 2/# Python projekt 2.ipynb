```python
##########################################NOTE FOR RADIM##############################################################
#Ahoj Radime,
#děkuji za hodnocení a zpětnou vazbu na první verzi mého druhého projektu.
#Do verze v1.1 jsem přidal kontrolu unikátnosti generovaného čísla a rozepsal trochu dokumentaci, jestli to lze v mém
#podání tak nazvat. Jádrová logika programu stojí na mém přehlédnutí požadavku unikátnosti jednotlivých cifer v tajném
#čísle, takže je asi komplikovanější, než musí být vzhledem ke zmíněné implementaci této kontroly. Na druhou stranu myslím,
#že aktuální fungování umožňuje po pár drobných úpravách měnit obtížnost hry a přídávat tak celkovou komplexitu výsledku.
#Nebo třeba dát možnost samotnému hráči si obtížnost zvolit (dle kombinace délky čísla, povolení duplikátů v tajném čísle apod.).
#Vzhledem k závěru tvého hodnocení a mému aktuálně nabitému programu jsem např. zmíněné výpočty bulls/cows neměnil.
#Děkuji za tvůj čas ale i motivační slova tvého hodnocení a měj se pěkně,
#Michal
#######################################################################################################################


#######################
###FOUNDATIONS SETUP###
#######################

import random

##########################################PROCESS COMMENT##############################################################
#while loop checks if the generated nr between 1000 and 9999 has unique digits or not
#it does so by comparing lengths of original str and set of it (sets keep only unique values, e.g. len(set("1111") = 1))
#hence only 4 unique values will be 4 digits long in such set
#######################################################################################################################
while True:
    number = str(random.randrange(1000, 9999))
    if len(set(number)) != len(number):
        #print(number, "not unique")
        continue
    else:
        #print(number, "UNIQUE")
        break

##########################################PROCESS COMMENT##############################################################
#welcome text setup and print, including variables setup, nothing that special here (as if the rest was...)
#######################################################################################################################
greetings = "Hello there!"
text = """I've generated a random 4 digit number for you.
The number is between 1000 and 9999 with UNIQUE digits.
If you get all numbers correct and in the right places, you win!
If not, you will get one bull for each number that is on the right spot
and/or one cow for each number that is in the generated number but you didnt
put it on the right spot.
Be careful, count matters! For example: generated nr = 1111, your nr = 4321 --> 0 bulls, 1 cow
Let's play a bulls and cows game.
I will keep the number of guesses it took to find the number."""
separator = "-" * (len(greetings) * 3 )
game_running = True
bulls = 0
cows = 0
used_guesses = list()

print(f"""
{separator}
{greetings}
{separator}
{text}
{separator}"""
, sep="\n")

##########################################PROCESS COMMENT##############################################################
#main "game tree" where all the magic happens by controlled infinite loop
#user enters his/hers guess which evaluated as shown and described below
#######################################################################################################################
while game_running:
    user_input = str(input("Enter a 4-digit number: "))

######################
###GUESS EVALUATION###
######################

    if user_input == "q":
        print("You decided to end the game. Goodbye ...")
        game_running = False
        break
    elif user_input != "q" and (len(user_input) != 4 or not user_input.isnumeric() or int(user_input) < 1000):
##########################################PROCESS COMMENT##############################################################
#this condition checks the possibilites where user inputs might not adhere to game rules
#that is done by leaving out the "leave option" (which is legit) and checking if the input is 4digits long, is a number etc
#this guess isnt counted as a guess (bcs of the continue keyword) and the show goes on
#######################################################################################################################
        print(f">>> {user_input}")
        print("Entered number doesnt have 4 digits, doesnt contain only numbers or is not between 1000 and 9999!")
        print(separator)
        continue
    elif user_input in used_guesses:
##########################################PROCESS COMMENT##############################################################
#this condition is player-friendly bcs it checks if the player havent guessed the current guess before
#it does so by looking into a list of previous guesses (which is updated each legit round)
#this guess isnt counted as a guess (bcs of the continue keyword) and the show goes on
#######################################################################################################################
        print(f"You have already tried {user_input} - try again something different!")
        print(separator)
        continue
    elif str(''.join(user_input)) == str(''.join(number)):
##########################################PROCESS COMMENT##############################################################
#this is a "winning" condition - if the guess matches the generated nr, the player wins
#as a small bonus the player is shown number of guesses he/she made and warm yet honest feedback
#######################################################################################################################
        used_guesses.append(user_input)
        if len(used_guesses) == 1:
            print(f"CONGRATS! You guessed the number: {user_input}! It took you only 1 guess!")
        else:
            print(f"CONGRATS! You guessed the number: {user_input}! It took you only {len(used_guesses)} guesses!")
        print(separator)
        if len(used_guesses) < 9:
            print("Damn, ur goooooood!")
        elif len(used_guesses) < 17:
            print("Well done, thats not bad at all...")
        elif len(used_guesses) < 26:
            print("There are other things you are good at, Im sure")
        elif len(used_guesses) >= 26:
            print("You better practice")
        break
    used_guesses.append(user_input)
    #here is the update of the used guesses list mentioned above

################################
###COWS AND BULLS CALCULATION###
################################

    user_input = list(user_input)
    number = list(number)
    bull_list = list()
    cow_list = list()
    rest_user_input = list()
    rest_number = list()
##########################################PROCESS COMMENT##############################################################
#logic of my solution goes as follows:
#   1) after initial guess evaluation above, the bulls/cows calculation is based bulls/cows lists and lists for remaining digits
#   2) firstly, for each digit of the input, the logic checks if on the same position of the digit is the same digit in the generated number
#       (as this solution is checking position by position in both data sources, the equality is tested to assess "bullness" of the digit)
#       if so, such digit is added to the bull list
#       if not, this digit is added to the list of remaining user input and the digit on the same position in the gen. number goes to equivavelent list
#       these lists are later used for cows calculation
#   3) then, for each digit of the remaining user input list, the logic checks if such digit has a friend in the rest gen. number list
#       (as cows only refer to digits in the gen. nr equal only in value, not position, group membership is tested)
#       if so, this digit is added to the cow list and removed from the list to be checked for the next digit
#       if not, nothing happens since such situation means no "formal output"
#######################################################################################################################
    for i in range(len(user_input)):
        if user_input[i] == number[i]:
            bull_list.append(int(user_input[i]))
        else:
            rest_user_input.append(user_input[i])
            rest_number.append(number[i])
    for i in range(len(rest_user_input)):
        if rest_user_input[i] in rest_number:
            rest_number.remove(str(rest_user_input[i]))
            cow_list.append(str(rest_user_input[i]))

###############################
###CALCULATIONN OUTPUT PRINT###
###############################

##########################################PROCESS COMMENT##############################################################
#here the continuous "progress" is given to the user (in case the guess adheres to game rules and isnt the gen. nr)
#the lengths of bulls and cows lists are used to determin the number of bulls/cows and print correct output
#######################################################################################################################
    print(f">>> {''.join(user_input)}")
    if len(bull_list) == 1:
        print(f"{len(bull_list)} bull,", end=' ')
    else:
        print(f"{len(bull_list)} bulls,", end=' ')
    if len(cow_list) == 1:
        print(f"{len(cow_list)} cow")
    else:
        print(f"{len(cow_list)} cows")
    print(separator)
```
