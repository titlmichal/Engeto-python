```python
min_temps, max_temps, rainfalls = [], [], []
with open(os.path.join("data", "weather.csv"), "r") as f:
    weather_reader = csv.reader(f, delimiter=",")
    for (i, row) in enumerate(weather_reader):
        if i == 0:
            continue
        min_temps.append(float(row[0]))
        max_temps.append(float(row[1]))
        rainfalls.append(float(row[2]))
```

```python
# import matplotlib.pyplot as plt
plt.plot(min_temps)
```

```python
fig, axs = plt.subplots(3, 1)
axs[0].plot(min_temps)
axs[1].plot(max_temps)
axs[2].plot(rainfalls)
```

```python
fig, axs = plt.subplots(3, 1, figsize=(15, 4))
axs[0].plot(min_temps, label="MinTemp", c="blue")
axs[1].plot(max_temps, label="MaxTemp", c="orange")
axs[2].plot(rainfalls, label="Rainfall", c="green")
fig.legend();
```
